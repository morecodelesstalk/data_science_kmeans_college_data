```python
from sklearn.preprocessing import StandardScaler, normalize

scaler = StandardScaler() 
scaled_df = scaler.fit_transform(df.drop('Private', axis = 1)) 
  
normalized = normalize(scaled_df)   
normalized_df = pd.DataFrame(normalized) 
```

### K-means cluster model

**Library imported from SciKit Learn.**

```python
from sklearn.cluster import KMeans
```

**Create a model with two clusters**

```python
kmeans = KMeans(n_clusters=2)
```

**fit normalized data**

```python
kmeans.fit(normalized_df)
```

```python
kmeans.predict
```

**check center vectors of cluster**

```python
centroids = kmeans.cluster_centers_
```

**plot centroids**

```python
centroids_x = centroids[:, 0]
centroids_y = centroids[:, 1]

plt.scatter(centroids_x, centroids_y, marker = "o", c=['red', 'blue'])
```

### Avaliation

We don't has any way to evaluating the groups without labels, however this is an exercice, let's go create label column for evaluating our model.

**Make a column called 'Cluster', with value 1 for private school and 0 for public school**

```python
def converter(cluster):
    if cluster == 'Yes':
        return 1
    else:
        return 0
```

```python
df['Cluster'] = df['Private'].apply(converter)
```

```python
df.head()
```

**print classification report with model fit without label column**

```python
from sklearn.metrics import confusion_matrix,classification_report

print(classification_report(df['Cluster'], kmeans.labels_))
```

**Now create a model and fit with dataset with label column, and check classification report**

```python
kmeans = KMeans(n_clusters = 2, random_state = 40)

kmeans.fit(df.drop('Private', axis = 1))

print(classification_report(df['Cluster'],kmeans.labels_))
```

### Get most optmal K value

```python
sse = {}

k = range(1,40)

for kclusters in k:
    kmeans1 = KMeans(n_clusters = kclusters)
    kmeans1.fit(df.drop('Private', axis=1))
    sse[kclusters] = kmeans1.inertia_
```

```python
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of clusters")
plt.ylabel("Sum of Squared distances - SSE")
plt.title("Clusteres per SSE - Algoritm KMeans no SKLearn")
plt.show()
```

```python
from sklearn.metrics import silhouette_score

silhouette_scores = [] 

k = range(2, 40)

for n_cluster in k:
    silhouette_scores.append( 
        silhouette_score(df.drop('Private', axis=1), KMeans(n_clusters = n_cluster).fit_predict(df.drop('Private', axis=1)))) 
```

```python
# Plotando um gráfico de barras para comparar os resultados
k = list(k)
plt.bar(k, silhouette_scores) 
plt.xlabel('Numbers of clusters', fontsize = 10) 
plt.ylabel('Silhouette Score', fontsize = 10) 
plt.show() 
```
